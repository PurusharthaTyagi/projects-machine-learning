```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(cancerX_df , cancerY_df , test_size = 0.2 , random_state = 5)
```

```python
min_train = x_train.min()
range_train = (x_train - min_train).max()
x_train_scaled = (x_train - min_train)/range_train
```

```python
min_test = x_test.min()
range_test = (x_test - min_test).max()
x_test_scaled = (x_test - min_test)/range_test
```

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(x_train_scaled, y_train)
```

```python
y_pred = classifier.predict(x_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix , classification_report , accuracy_score
```

```python
cm = confusion_matrix(y_test , y_pred)
```

```python
cm
```

```python
sns.heatmap(data = cm , annot= True)
```

```python
classification_report(y_test , y_pred)
```

```python
accuracy_score(y_test , y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
param = [{'C' : [0.1 ,0.001 , 1 , 10, 100 ,1000 ] , 'kernel' : ['rbf'] , 'gamma' : [.0001,.001 ,.01,.1,1,10] , }]
```

```python
grid =GridSearchCV(SVC() , param , refit =True , verbose = 4)
```

```python
grid.fit(x_train_scaled , y_train)
```

```python
grid.best_params_
```

```python
grid_predictions=grid.predict(x_test_scaled)
```
